```python
import pandas as pd
import numpy as np
```

```python
vykony = pd.read_csv("./data/vykazy/vyk_23_vykony_new.csv", sep=";", encoding="windows-1250")
```

```python
vykony_pivot = pd.get_dummies(vykony.set_index('CDOKL')['KOD']).groupby('CDOKL').sum()
vykony_pivot
co_occurrence_matrix =  np.dot(vykony_pivot.T, vykony_pivot)
# Normalize each row by dividing by row sums
row_sums = co_occurrence_matrix.sum(axis=1)
```

```python
row_sums
```

```python
row_sums[row_sums == 0] = 1  # Avoid division by zero
co_occurrence_matrix = co_occurrence_matrix / row_sums[:, np.newaxis]


# normalize values in matrix across rows
np.fill_diagonal(co_occurrence_matrix, 0)
```

```python
# co_occurrence_matrix
co_occurrence_df = pd.DataFrame(co_occurrence_matrix, index=vykony_pivot.columns, columns=vykony_pivot.columns)
```

```python
co_occurrence_df
```

```python
# Get rows that have at least one positive value
co_occurrence_df = co_occurrence_df[(co_occurrence_df > 0).any(axis=1)]
```
